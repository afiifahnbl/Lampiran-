```python
# Langkah 1: Import library yang dibutuhkan
import pandas as pd
import matplotlib.pyplot as plt

# Langkah 2: Baca file yang berisi data Chain A
file_path = 'chaina.txt'  # Pastikan nama file sudah sesuai setelah diunggah
data = pd.read_csv(file_path, delimiter='\t')

# Langkah 3: Pilih kolom kode protein dan nomor residu dari file
kode_protein_columns = ['5v3w_a', '5v3y_a', '5v3z_a', '5v3x_a', '5v40_a', '5v41_a', '5v42_a']
residue_numbers = data['Unnamed: 1']  # Kolom dengan nomor residu

# Langkah 4: Filter untuk residu ASN di posisi 1607 pada kode protein 5v3z_a
target_position = (residue_numbers == 1607)  # Mendapatkan posisi 1607
target_amino_acid = data['5v3z_a'][target_position].values[0] == 'ASN'

# Langkah 5: Plot semua nomor residu dan asam amino dari awal hingga akhir untuk setiap kode protein
plt.figure(figsize=(10, 12))
for col in kode_protein_columns:
    plt.scatter([col] * len(residue_numbers), residue_numbers, color='gray', alpha=0.1)  # Tampilkan semua residu sebagai titik abu-abu ringan

# Menambahkan tanda khusus untuk ASN di kode protein 5v3z_a pada posisi 1607
if target_amino_acid:
    plt.scatter('5v3z_a', 1607, color='red', s=100, label="ASN pada Posisi 1607 di 5v3z_a")

# Menambahkan keterangan nomor residu dan asam amino untuk ASN di luar plot
plt.text(1.05, 1607, f"ASN (1607)", ha='left', va='center', fontsize=10, color='black')

# Pengaturan grafik
plt.title("Perbedaan Sequence di Residu 1607 ASN pada Kode Protein 5v3z_a")
plt.xlabel("Kode Protein")
plt.ylabel("Nomor Residu dan Asam Amino")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)

# Langkah 6: Simpan file sebagai PDF dengan resolusi 600 dpi
plt.savefig("plot_protein_sequence_difference_1607ASN.pdf", format='pdf', dpi=600, bbox_inches='tight')

# Menampilkan plot
plt.show()
```
