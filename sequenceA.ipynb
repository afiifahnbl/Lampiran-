```python
# Langkah 1: Import library yang dibutuhkan
import pandas as pd
import matplotlib.pyplot as plt

# Langkah 2: Baca file yang berisi data chain
file_path = 'new1.txt'  # Ganti dengan path file Anda setelah diunggah di Google Colab
data = pd.read_csv(file_path, delimiter='\t')

# Langkah 3: Pilih kolom untuk Chain A (misalnya, "5v3z_a") dan Chain B untuk dibandingkan
chain_a = data['5v3z_a'].astype(str)  # Mengonversi ke string untuk menghindari error
chain_b = data['5v3w_a'].astype(str)  # Mengonversi ke string
residue_numbers = data['Unnamed: 1']  # Sesuaikan dengan kolom nomor residu

# Langkah 4: Identifikasi semua residu di Chain A dengan nomor residu
amino_acids_a = chain_a
residue_numbers_a = residue_numbers

# Langkah 5: Tandai hanya perbedaan pada urutan 1607 dengan kode protein 5v3z_a
# Filter untuk residu 1607 yang memiliki ASN di Chain A
specific_differences = (residue_numbers == 1607) & (chain_a == 'ASN')

# Data untuk plotting
different_residues_a = chain_a[specific_differences]
different_residue_numbers_a = residue_numbers[specific_differences]

# Langkah 6: Plot hasil
plt.figure(figsize=(12, 8))
plt.scatter(amino_acids_a, residue_numbers_a, color='blue', alpha=0.6, label="Asam Amino Chain A")

# Menambahkan tanda untuk residu yang berbeda pada posisi 1607 ASN
plt.scatter(different_residues_a, different_residue_numbers_a, color='red', alpha=0.8, label="Perbedaan ASN di Posisi 1607")

# Menambahkan keterangan nomor residu dan asam amino untuk perbedaan ini
for i in range(len(different_residues_a)):
    plt.text(different_residues_a.iloc[i], different_residue_numbers_a.iloc[i],
             f"{different_residues_a.iloc[i]} ({different_residue_numbers_a.iloc[i]})",
             ha='center', va='bottom', fontsize=8, color='black', rotation=45)

# Pengaturan grafik
plt.title("Plot Semua Asam Amino di Chain A dengan Nomor Residu dan Perbedaan di Posisi 1607")
plt.xlabel("Jenis Asam Amino di Chain A")
plt.ylabel("Nomor Residu")
plt.xticks(rotation=90)  # Memutar label asam amino agar lebih mudah dibaca
plt.legend()
plt.grid(True)

# Langkah 7: Simpan file sebagai PDF dengan resolusi 600 dpi
plt.savefig("plot_chain_a_1607_difference.pdf", format='pdf', dpi=600)

# Menampilkan plot
plt.show()
```
